```python
import pandas as pd

df = pd.read_csv("../data/processed/clean_housing_subset.csv")
df.head()

data_dict = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype)for c in df.columns],
    "missing_count": [int(df[c].isna().sum()) for c in df.columns],
    "missing_pct": [round(df[c].isna().mean() * 100, 1) for c in df.columns],
    "example_value": [df[c].dropna().iloc[0] if df[c].dropna().shape[0] > 0 else None for c in df.columns]
})
data_dict 

data_dict.to_csv("../data/processed/data_dictionary.csv", index=False)
print("Saved: ../data/processed/data_dictionary.csv")
```
